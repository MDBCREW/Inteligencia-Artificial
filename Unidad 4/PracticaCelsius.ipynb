```python
import tensorflow as tf
import numpy as np
```

```python
celsius = np.array([-40, -25, -10, 0, 10, 25, 35, 50], dtype=float)
farenheit = np.array([-40, -13, 14, 32, 50, 77, 95, 122], dtype=float)
```

```python
capa_1 = tf.keras.layers.Dense(units=3, input_shape=[1])
capa_2 = tf.keras.layers.Dense(units=3)
salida = tf.keras.layers.Dense(units=1)
modelo = tf.keras.Sequential([capa_1, capa_2, salida])
```

```python
modelo.compile(
    optimizer = tf.keras.optimizers.Adam(0.5),
    loss = 'mean_squared_error'
) 
```

```python
print ("A punto de entrenar mi modelo... ")
historia = modelo.fit(celsius, farenheit, epochs=1000)
print ("Modelo ")
```
